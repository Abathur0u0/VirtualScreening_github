```python
import pandas as pd
# 讀取 Excel 文件並轉換為 DataFrame
df = pd.read_excel('CNSData(940).xlsx')
# 打印前五個分子的ID、Name、SMILES、Output
test_df = df.head()
test_df
```

```python
from padelpy import from_smiles

descriptors_list = []
for smiles in test_df['smiles']:
    desc = from_smiles(smiles)
    descriptors_list.append(desc)

# 將描述符列表轉換為 DataFrame
descriptors_df = pd.DataFrame(descriptors_list)

# 將分子的 ID、Name、SMILES 和 Output 列與描述符 DataFrame 拼接在一起
result_df = pd.concat([test_df, descriptors_df], axis=1)

# 打印結果 DataFrame
result_df.head()
```

```python
from padelpy import from_smiles

descriptors_list = []
for smiles in df['smiles']:
    desc = from_smiles(smiles)
    descriptors_list.append(desc)

# 將描述符列表轉換為 DataFrame
descriptors_df = pd.DataFrame(descriptors_list)

# 將分子的 ID、Name、SMILES 和 Output 列與描述符 DataFrame 拼接在一起
result_df = pd.concat([df, descriptors_df], axis=1)

# 打印結果 DataFrame
result_df.head()
```

```python
# 找出在訓練資料集中所有數值相同的列
constant_columns = result_df.columns[result_df.nunique() == 1].tolist()

# 在訓練資料集中刪除這些列
dropped_result_df = result_df.drop(columns=constant_columns)

# constant_columns 變數中包含了在訓練資料集中所有數值相同的列的名稱
print("在訓練資料集中所有數值相同的列的名稱：", constant_columns)
dropped_result_df.head()
```

```python
# 將數據保存為 xlsx 文件
result_df.to_excel('Data_with_1879_desc.xlsx', index=False)
dropped_result_df.to_excel('OUTPUT.xlsx', index=False)
```
